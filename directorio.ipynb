```python
spacy.prefer_gpu()
nlp = spacy.load("es_core_news_md")
```

```python
joined =" ".join(descriptions)
doc = nlp(joined)
```

```python
sentence1 = list(doc.sents)[12]
print(sentence1)
displacy.render(sentence1, style="dep")
```

```python
displacy.render(doc, style="ent")
```

```python
#https://stackoverflow.com/questions/11184751/how-to-do-loop-in-python-map
word = "hombre"
vector = nlp.vocab.vectors[nlp.vocab.strings[word]]
vector = cp.asarray([vector])  
ms = nlp.vocab.vectors.most_similar(vector, n=10)
words = [nlp.vocab.strings[w] for w in ms[0][0]]
distances = ms[2]

print("Most similar words:", words)
print("Distances:", distances)
```
